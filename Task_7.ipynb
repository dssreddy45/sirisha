# Transformers, Attention Mechanism & Embeddings (Foundation of LLMs)

Foundation of Large Language Models (LLMs)

Modern LLMs such as GPT, BERT, and LLaMA are built on three key concepts:

Embeddings,
Attention Mechanism,
Transformers

1. Embeddings

Embeddings convert words or tokens into numerical vectors that capture their meaning and relationships. Similar words have similar vector representations, helping the model understand semantics rather than just text.

Example:
King → [0.8, 0.2, ...]
Queen → [0.79, 0.25, ...]

2. Attention Mechanism

Attention allows a model to focus on the most relevant words in a sentence when processing or generating text. It helps capture context and relationships between words, even when they are far apart.

Example:
In "The dog chased the ball because it was fast," attention helps determine what "it" refers to.

3. Transformers

Transformers are deep learning architectures built around self-attention mechanisms. They process all words in parallel, making training faster and enabling better handling of long-range dependencies. Modern LLMs are based on Transformer architectures.

Flow:

Input Text
    ↓
Embeddings
    ↓
Self-Attention
    ↓
Transformer Layers
    ↓
Output Prediction

# Traditional NLP methods

Traditional NLP converts text into numerical features using Tokenization, Bag of Words, or TF-IDF and then applies machine learning algorithms for prediction. While effective for simple tasks, it lacks deep contextual understanding compared to modern Transformer-based models.

# Modern LLM approach:

Modern LLMs process text by converting tokens into embeddings, passing them through Transformer layers that use attention mechanisms to understand context and relationships, and then generating intelligent outputs one token at a time. This architecture enables powerful language understanding and generation.

# Task 2: Implement Basic Text Embeddings

```python
# ── Install Libraries ──────────────────────────────────────
!pip install sentence-transformers numpy scikit-learn -q

# ── Imports ────────────────────────────────────────────────
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer

# ── 1. Load Model ──────────────────────────────────────────
print("🔄 Loading model...")
model = SentenceTransformer('all-MiniLM-L6-v2')
print("✅ Model Loaded!\n")

# ── 2. Define Sentences ────────────────────────────────────
sentences = [
    "AI is changing the world.",
    "Artificial Intelligence is transforming industries.",
    "I love playing cricket."
]

# ── 3. Generate Embeddings ─────────────────────────────────
embeddings = model.encode(sentences)

print("✅ Embeddings Generated!")
print(f"   Embedding Dimensions : {embeddings.shape[1]}")
print(f"   Total Sentences      : {embeddings.shape[0]}\n")

# ── 4. Display Embedding Vectors ───────────────────────────
print("="*55)
print("  EMBEDDING VECTORS (first 5 values)")
print("="*55)
for i, sentence in enumerate(sentences):
    print(f"\n  Sentence {i+1}: {sentence}")
    print(f"  Vector    : {embeddings[i][:5].round(4)} ...")
    print(f"  Dimensions: {len(embeddings[i])}")

# ── 5. Calculate Cosine Similarity ─────────────────────────
sim_1_2 = cosine_similarity([embeddings[0]], [embeddings[1]])[0][0]
sim_1_3 = cosine_similarity([embeddings[0]], [embeddings[2]])[0][0]
sim_2_3 = cosine_similarity([embeddings[1]], [embeddings[2]])[0][0]

print("\n" + "="*55)
print("  COSINE SIMILARITY SCORES")
print("="*55)
print(f"  Sentence 1 vs Sentence 2 : {sim_1_2:.4f} ✅ (Similar)")
print(f"  Sentence 1 vs Sentence 3 : {sim_1_3:.4f} ❌ (Different)")
print(f"  Sentence 2 vs Sentence 3 : {sim_2_3:.4f} ❌ (Different)")

# ── 6. Visual Bar Chart ────────────────────────────────────
print("\n" + "="*55)
print("  SIMILARITY BAR CHART")
print("="*55)
pairs = [
    ("S1 vs S2", sim_1_2),
    ("S1 vs S3", sim_1_3),
    ("S2 vs S3", sim_2_3),
]
for label, score in pairs:
    bar = "█" * int(score * 40)
    print(f"  {label} : {score:.4f}  {bar}")

# ── 7. Explanation ─────────────────────────────────────────
print("\n" + "="*55)
print("  WHY SENTENCES 1 AND 2 ARE CLOSER")
print("="*55)
print("""
  Sentence 1: AI is changing the world.
  Sentence 2: Artificial Intelligence is transforming industries.
  Sentence 3: I love playing cricket.

  Reason 1 - Same Topic:
  Sentences 1 and 2 both talk about AI and its
  impact on the world. Same topic = similar vectors.

  Reason 2 - Synonym Words:
  AI = Artificial Intelligence
  changing = transforming
  world = industries
  The model understands these are the same concepts.

  Reason 3 - Sentence 3 is Unrelated:
  Cricket is a sport with no connection to AI.
  The embedding vector points in a completely
  different direction in the vector space.

  Cosine Similarity Interpretation:
  1.0  = Identical meaning
  0.8+ = Very similar
  0.5  = Somewhat related
  0.0  = Completely unrelated
""")
```

# Task 3: Build Semantic Search Engine

```python
# Install sentence-transformers (run once)
!pip install -q sentence-transformers

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# 1. Load a pretrained embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# 2. Dataset
documents = [
    "Python is used for AI development",
    "Machine learning helps computers learn patterns",
    "Transformers power modern LLMs",
    "Cloud computing provides scalable infrastructure",
    "Databases store application data"
]

# 3. Generate embeddings and store them in memory
doc_embeddings = model.encode(documents)

def semantic_search(query, top_k=3):
    # Embed the query
    query_embedding = model.encode([query])

    # Calculate cosine similarity between query and all documents
    scores = cosine_similarity(query_embedding, doc_embeddings)[0]

    # Rank documents by similarity score (highest first)
    ranked_indices = np.argsort(scores)[::-1][:top_k]

    return [(documents[i], float(scores[i])) for i in ranked_indices]

# 4. Accept a user query and run the search
query = input("Enter your query: ")
results = semantic_search(query, top_k=3)

print(f"\nTop {len(results)} results for: '{query}'\n")
for rank, (doc, score) in enumerate(results, start=1):
    print(f"{rank}. {doc}  (similarity: {score:.4f})")
```

# Task 5: Tokenization Experiment

```python
# Install required library (run once)
!pip install -q transformers

from transformers import AutoTokenizer

# Load a pretrained Hugging Face tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Sample texts for comparison
short_sentence = "Tokenization breaks text into smaller units."

long_paragraph = (
    "Tokenization is the process of splitting text into smaller pieces, "
    "called tokens, before feeding it into a machine learning model. "
    "Modern transformer models like BERT and GPT use subword tokenization, "
    "which breaks rare or unknown words into smaller, more common pieces. "
    "This approach helps the model handle a much larger vocabulary than "
    "word-level tokenization while keeping the total number of unique "
    "tokens manageable. As a result, even words the model has never seen "
    "during training can often still be represented as a sequence of "
    "familiar subword units."
)

code_snippet = """def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)"""

samples = {
    "Short Sentence": short_sentence,
    "Long Paragraph": long_paragraph,
    "Programming Code": code_snippet
}

# Tokenize, convert to IDs, decode back to text
for label, text in samples.items():
    print(f"===== {label} =====")
    tokens = tokenizer.tokenize(text)
    token_ids = tokenizer.convert_tokens_to_ids(tokens)
    decoded_text = tokenizer.decode(token_ids)

    print(f"Original text: {text}")
    print(f"Tokens: {tokens}")
    print(f"Token IDs: {token_ids}")
    print(f"Decoded text: {decoded_text}")
    print(f"Token count: {len(tokens)}\n")

# Compare token counts
print("===== Token Count Comparison =====")
for label, text in samples.items():
    print(f"{label}: {len(tokenizer.tokenize(text))} tokens")
```

```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
import torch

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-cased-distilled-squad")
model = AutoModelForQuestionAnswering.from_pretrained("distilbert-base-cased-distilled-squad")

context = (
    "Python is a programming language created by Guido van Rossum. "
    "It is widely used in AI, web development, automation, and data science."
)

questions = [
    "Who created Python?",
    "Where is Python used?"
]

for question in questions:
    inputs = tokenizer(question, context, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)

    start = torch.argmax(outputs.start_logits)
    end = torch.argmax(outputs.end_logits) + 1
    answer = tokenizer.convert_tokens_to_string(
        tokenizer.convert_ids_to_tokens(inputs["input_ids"][0][start:end])
    )

    print(f"Question : {question}")
    print(f"Answer   : {answer}\n")
```

# Task 8: Build an LLM Foundation Report

Large Language Models (LLMs) like ChatGPT have transformed the way humans interact with machines. They can answer questions, write code, summarize documents, translate languages, and hold conversations that feel remarkably human. But how do they actually work under the hood? This report breaks down five foundational concepts that explain how LLMs understand language and generate responses.

1. How ChatGPT Understands a Question
When you type a question into ChatGPT, it does not read it the way a human does. Instead, it processes the text through a series of mathematical transformations.
Step 1 — Tokenization: The first thing ChatGPT does is break your input into tokens. A token is not always a full word — it can be a word, a part of a word, or punctuation. For example "I love AI" becomes ["I", "love", "AI"].
Step 2 — Embedding: Each token is converted into a dense numerical vector called an embedding. The model also adds positional encodings that tell the model where in the sentence each token appears.
Step 3 — Transformer Layers: The embeddings pass through many transformer layers. In each layer, self-attention allows every token to look at every other token and decide how much to pay attention to it.
Step 4 — Probability Distribution: After all layers, the model produces a probability score for every token in the vocabulary — representing how likely each token is to come next.
Step 5 — Autoregressive Generation: ChatGPT generates one token at a time. After picking the first token, it appends it to the input and repeats the process until the full response is generated.

2. How Embeddings Represent Meaning
An embedding is a list of numbers (a vector) that represents text in a high-dimensional mathematical space. Semantically similar texts end up with vectors that are close together, while unrelated texts are far apart.
Embeddings are not hand-crafted — they are learned during training. Word2Vec (2013) introduced dense vectors where similar words cluster together. That is why the famous equation "king − man + woman ≈ queen" holds in embedding space.
Modern transformer models like BERT produce contextual embeddings — the same word gets a different vector depending on the sentence. "Bank" in "river bank" gets a different embedding than "bank" in "savings bank."
Embeddings capture synonyms, analogies, sentiment, and topic clusters. They are the bridge between human language and machine mathematics.

3. How Attention Helps Models Understand Context
Attention is the most important innovation in modern NLP. It allows the model to understand which parts of the input are relevant to each other.
For each token, the model computes three vectors — a Query (Q), a Key (K), and a Value (V). The attention score between two tokens is the dot product of one token's Query and another's Key, passed through softmax. The output is a weighted sum of all Value vectors.
Example: In "The trophy didn't fit in the suitcase because it was too big" — what does "it" refer to? Attention allows the model to compute high scores between "it" and "trophy" based on context, correctly resolving the reference.
Transformers use multi-head attention — several attention operations run in parallel, each learning different types of relationships like subject-verb agreement or coreference.

4. Why Transformers Replaced RNNs
Before transformers, Recurrent Neural Networks (RNNs) and LSTMs dominated NLP. They processed text sequentially — one token at a time — maintaining a hidden state.
Problems with RNNs:

Vanishing gradients: error signals shrink over long sequences, making it hard to learn long-range dependencies.
Sequential processing: tokens must be processed one by one, making training slow and hard to parallelize.
Limited context: information from early in a long sequence gets diluted by the end.

Why transformers win:

Full parallelization: all tokens are processed simultaneously using matrix multiplications.
Direct long-range dependencies: any token can attend to any other token regardless of distance.
Scalability: transformers scale predictably with more data and parameters, leading to GPT-3, GPT-4, and beyond.

The 2017 paper "Attention Is All You Need" introduced transformers and made RNNs effectively obsolete for large-scale NLP almost overnight.

5. How LLMs Generate the Next Word
After the input passes through all transformer layers, the final hidden state is projected through a linear layer to produce logits — one number per vocabulary token. Softmax converts these into probabilities.
Sampling strategies:

Greedy decoding: always pick the highest-probability token. Fast but repetitive.
Temperature sampling: high temperature = more creative output. Low temperature = more predictable output.
Top-k sampling: only sample from the k most likely tokens.
Top-p (nucleus) sampling: sample from the smallest set of tokens whose cumulative probability exceeds p.

The model generates text in a loop — pick a token, append it to the input, repeat — until an end token is produced. This is why inference on large models is computationally expensive.
