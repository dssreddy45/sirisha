# Natural Language Processing (NLP) Fundamentals

NLP (Natural Language Processing) is a field of AI that enables computers to understand, interpret, process, and generate human language (such as English, Telugu, Hindi, etc.) in a meaningful way.

# NLP Pipeline

**1. Raw Text**
Original text collected from users, documents, messages, or websites.

**2. Cleaning**
Remove unwanted characters, symbols, punctuation, and convert text into a usable format.

**3. Tokenization**
Split text into smaller parts such as words or sentences.

**4. Stop Word Removal**
Remove common words (like *is, the, and*) that add little meaning.

**5. Stemming / Lemmatization**
Convert words into their base or root form.

**6. Vectorization**
Convert text into numerical values so machine learning models can process it.

**7. Model Training**
Train the machine learning model to learn patterns and make predictions from text data.

# Task 2: Text Cleaning

Text Cleaning: Text cleaning is the process of removing unnecessary or unwanted information from raw text data to make it clean and ready for analysis or machine learning. It includes tasks like converting text to lowercase, removing special characters, URLs, emails, phone numbers, and extra spaces.

```python
import re

# Input text
text = """
Hello Everyone!!!

Welcome to AI/ML Training.

Contact us at support@company.com

Visit https://company.com

Call 9876543210
"""

# Convert to lowercase
text = text.lower()

# Remove URLs
text = re.sub(r'https?://\S+|www\.\S+', '', text)

# Remove email addresses
text = re.sub(r'\S+@\S+', '', text)

# Remove phone numbers
text = re.sub(r'\b\d{10}\b', '', text)

# Remove special characters
text = re.sub(r'[^a-zA-Z\s]', '', text)

# Remove extra spaces
text = re.sub(r'\s+', ' ', text).strip()

# Output cleaned text
print("Cleaned Text:")
print(text)
```

# Task 3: Tokenization

**Tokenization:** Tokenization is the process of splitting text into smaller units called tokens, such as words, sentences, or characters, for text analysis and NLP tasks.


```python
# Install and import NLTK
!pip install nltk

import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize, sent_tokenize

# Input sentence
text = "Artificial Intelligence is transforming software development rapidly."

# Word Tokenization
word_tokens = word_tokenize(text)

# Sentence Tokenization
sentence_tokens = sent_tokenize(text)

# Display tokens
print("Word Tokens:")
print(word_tokens)

print("\nSentence Tokens:")
print(sentence_tokens)
```

# Task 4: Stop Word Removal

Stop Word Removal: Stop word removal is the process of removing common words (such as is, the, and, in) from text to keep only meaningful words for NLP analysis.

```python
# Install and import NLTK
!pip install nltk

import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Input text
text = "Artificial Intelligence is one of the most important technologies in the world."

# Tokenize text
tokens = word_tokenize(text)

# Load stop words
stop_words = set(stopwords.words('english'))

# Remove stop words
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

# Convert tokens back to text
processed_text = " ".join(filtered_tokens)

# Display output
print("Original Text:")
print(text)

print("\nProcessed Text:")
print(processed_text)
```

# Task 5: Stemming & Lemmatization

Stemming & Lemmatization: Stemming reduces words to their root form by removing suffixes, while lemmatization converts words into their meaningful dictionary base form.

```python
# Install and import NLTK
!pip install nltk

import nltk
nltk.download('wordnet')

from nltk.stem import PorterStemmer, WordNetLemmatizer

# Input words
words = [
    "running",
    "playing",
    "studies",
    "better",
    "cars"
]

# Initialize
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

print("Word\t\tStemming\tLemmatization")
print("-" * 45)

# Compare results
for word in words:
    stem = stemmer.stem(word)
    lemma = lemmatizer.lemmatize(word)
    print(f"{word}\t\t{stem}\t\t{lemma}")
```

# Task 6: Text Vectorization

Text Vectorization (Definition): Text vectorization is the process of converting text into numerical format so machine learning models can understand and process text data.

```python
# Install scikit-learn
!pip install scikit-learn

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Dataset
documents = [
    "AI is transforming businesses",
    "Machine learning powers AI",
    "Deep learning is a subset of AI",
    "AI applications are growing rapidly"
]

# ---------------- Bag of Words ----------------
bow = CountVectorizer()
bow_matrix = bow.fit_transform(documents)

print("=== Bag of Words ===")
print("Vocabulary:")
print(bow.vocabulary_)

print("\nMatrix Shape:")
print(bow_matrix.shape)

print("\nFeature Names:")
print(bow.get_feature_names_out())


# ---------------- TF-IDF ----------------
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(documents)

print("\n=== TF-IDF ===")
print("Vocabulary:")
print(tfidf.vocabulary_)

print("\nMatrix Shape:")
print(tfidf_matrix.shape)

print("\nFeature Names:")
print(tfidf.get_feature_names_out())
```

# Task 7: Sentiment Analysis

Sentiment Analysis (Definition): Sentiment analysis is the process of identifying whether text expresses a positive, negative, or neutral opinion.

```python
# Install TextBlob
!pip install textblob

from textblob import TextBlob

# Dataset
reviews = [
    "This course is amazing",
    "The training is excellent",
    "The project is difficult",
    "The application is terrible"
]

print("Sentiment Analysis Report")
print("-" * 40)

positive = 0
negative = 0
neutral = 0

for review in reviews:
    polarity = TextBlob(review).sentiment.polarity

    if polarity > 0:
        sentiment = "Positive"
        positive += 1
    elif polarity < 0:
        sentiment = "Negative"
        negative += 1
    else:
        sentiment = "Neutral"
        neutral += 1

    print(f"Review: {review}")
    print(f"Sentiment: {sentiment}\n")

# Summary Report
print("Final Report")
print("-" * 20)
print("Positive:", positive)
print("Negative:", negative)
print("Neutral:", neutral)
```

```python
import re, collections

STOP_WORDS = {
    "a","an","the","and","or","but","in","on","at","to","for","of","with","by",
    "from","is","are","was","were","be","been","have","has","had","do","does",
    "did","will","would","could","should","may","might","can","not","no","it",
    "its","i","me","my","we","you","your","he","she","they","their","them","s",
    "t","using","used","use","also","more","such","own","same","very","just",
}

resume_text = """
John Doe
Software Engineer
Skills: Python, Java, SQL, Machine Learning, Data Analysis
Experience: 3 years at TechCorp, developed REST APIs using Django and Flask
Education: B.Tech Computer Science 2020
Certifications: AWS, TensorFlow, Docker, Kubernetes
"""

text = resume_text.lower()
text = re.sub(r"[^a-z\s]", " ", text)
text = re.sub(r"\s+", " ", text).strip()
tokens = [t for t in text.split() if len(t) >= 2 and t not in STOP_WORDS]
keywords = collections.Counter(tokens).most_common(20)

print("TOP 20 KEYWORDS")
print("-" * 35)
for rank, (word, freq) in enumerate(keywords, 1):
    bar = "█" * freq
    print(f"{rank:<3} {word:<20} {freq}  {bar}")
```

# Bonus Challenge

```python
import re, collections

# ── Stop Words ──────────────────────────────────────────────
STOP_WORDS = {
    "a","an","the","and","or","but","in","on","at","to","for","of","with","by",
    "from","is","are","was","were","be","been","have","has","had","do","does",
    "did","will","would","could","should","may","might","can","not","no","it",
    "its","i","me","my","we","you","your","he","she","they","their","them","s",
    "t","using","used","use","also","more","such","own","same","very","just",
    "im","ive","id","ill","dont","cant","wont","isnt","wasnt","didnt","like",
    "get","got","go","going","know","think","want","need","make","tell","say",
}

# ── Simple Lemmatizer (rule-based) ──────────────────────────
LEMMA_RULES = {
    "running":"run","runs":"run","runner":"run",
    "learning":"learn","learned":"learn","learns":"learn",
    "working":"work","worked":"work","works":"work",
    "helping":"help","helped":"help","helps":"help",
    "playing":"play","played":"play","plays":"play",
    "talking":"talk","talked":"talk","talks":"talk",
    "sending":"send","sent":"send","sends":"send",
    "buying":"buy","bought":"buy","buys":"buy",
    "fixing":"fix","fixed":"fix","fixes":"fix",
    "building":"build","built":"build","builds":"build",
    "writing":"write","wrote":"write","writes":"write",
    "reading":"read","reads":"read",
    "issues":"issue","errors":"error","messages":"message",
    "users":"user","files":"file","apps":"app","chats":"chat",
    "problems":"problem","solutions":"solution","questions":"question",
    "services":"service","features":"feature","updates":"update",
    "faster":"fast","slower":"slow","better":"good","worse":"bad",
}

def lemmatize(tokens):
    result = []
    for t in tokens:
        if t in LEMMA_RULES:
            result.append(LEMMA_RULES[t])
        elif t.endswith("ing") and len(t) > 5:
            result.append(t[:-3])
        elif t.endswith("tion"):
            result.append(t[:-3] + "e")
        elif t.endswith("ed") and len(t) > 4:
            result.append(t[:-2])
        elif t.endswith("ly") and len(t) > 4:
            result.append(t[:-2])
        elif t.endswith("es") and len(t) > 4:
            result.append(t[:-2])
        elif t.endswith("s") and len(t) > 4:
            result.append(t[:-1])
        else:
            result.append(t)
    return result

# ── 1. Clean Text ───────────────────────────────────────────
def clean_text(text):
    text = text.lower()
    text = re.sub(r"https?://\S+|www\.\S+", " ", text)
    text = re.sub(r"\S+@\S+\.\S+", " ", text)
    text = re.sub(r"[^a-z\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

# ── 2. Tokenize ─────────────────────────────────────────────
def tokenize(text):
    return [t for t in text.split() if len(t) >= 2]

# ── 3. Remove Stop Words ────────────────────────────────────
def remove_stopwords(tokens):
    return [t for t in tokens if t not in STOP_WORDS]

# ── 4. Lemmatize ────────────────────────────────────────────
def lemmatize_tokens(tokens):
    return lemmatize(tokens)

# ── 5. Generate Keywords ────────────────────────────────────
def generate_keywords(tokens, top_n=10):
    return collections.Counter(tokens).most_common(top_n)

# ── 6. Export Report ────────────────────────────────────────
def export_report(messages, keywords, stats):
    report = []
    report.append("=" * 55)
    report.append("       AI CHAT PREPROCESSOR — REPORT")
    report.append("=" * 55)
    report.append(f"  Total Messages    : {stats['messages']}")
    report.append(f"  Total Tokens      : {stats['total_tokens']}")
    report.append(f"  After Stop Words  : {stats['filtered_tokens']}")
    report.append(f"  After Lemmatize   : {stats['lemma_tokens']}")
    report.append(f"  Unique Keywords   : {stats['unique']}")
    report.append("=" * 55)
    report.append("  TOP KEYWORDS")
    report.append("-" * 55)
    max_freq = keywords[0][1] if keywords else 1
    for rank, (word, freq) in enumerate(keywords, 1):
        bar = "█" * int((freq / max_freq) * 30)
        report.append(f"  {rank:<3} {word:<18} {freq:>3}  {bar}")
    report.append("=" * 55)

    # Print to screen
    print("\n".join(report))

    # Save to file
    with open("chat_report.txt", "w") as f:
        f.write("\n".join(report))
    print("\n✅ Report saved to chat_report.txt")

# ── CHAT MESSAGES INPUT ─────────────────────────────────────
chat_messages = [
    "I am having trouble logging into my account",
    "The app keeps crashing when I open it",
    "Can you help me reset my password please",
    "I need help with payment not working",
    "The app is running very slow today",
    "How do I update my profile settings",
    "I cannot send messages to other users",
    "My subscription is not working properly",
    "Can you fix the error in my account",
    "I want to cancel my subscription plan",
]

# ── PIPELINE ────────────────────────────────────────────────
all_tokens = []
all_lemmas = []

print("PROCESSING MESSAGES...\n")
for i, msg in enumerate(chat_messages, 1):
    cleaned  = clean_text(msg)
    tokens   = tokenize(cleaned)
    filtered = remove_stopwords(tokens)
    lemmas   = lemmatize_tokens(filtered)
    all_tokens.extend(filtered)
    all_lemmas.extend(lemmas)
    print(f"  [{i}] {msg}")
    print(f"      Keywords: {lemmas}\n")

keywords = generate_keywords(all_lemmas, top_n=10)
stats = {
    "messages"       : len(chat_messages),
    "total_tokens"   : len(all_tokens),
    "filtered_tokens": len(all_tokens),
    "lemma_tokens"   : len(all_lemmas),
    "unique"         : len(set(all_lemmas)),
}

export_report(chat_messages, keywords, stats)
```
