# Task 2: Build AI API Service

Build AI API Service :

An AI API Service is a backend application that exposes AI functionalities through REST APIs. It allows clients such as web applications, mobile apps, and other services to interact with AI models using HTTP requests.

Endpoints Implemented:

1.Health Check (GET /health):

Verifies that the API service is running correctly.
Returns the service status and timestamp.

2.Chat Endpoint (POST /chat):

Accepts a user message.
Processes the request and returns an AI-generated response.

3.Embedding Endpoint (POST /embeddings):

Converts input text into numerical vector representations (embeddings).
Used for semantic search, clustering, and recommendation systems.

4.Document Search Endpoint (POST /search):

Searches documents based on a query.
Returns the most relevant results.

Key Features:

1.Validation:

Uses Pydantic models to validate incoming request data.

2.Error Handling:

Handles invalid requests and server errors using HTTP exceptions.

3.Logging:

Records API activities, requests, and errors for monitoring and debugging.

Benefits:

1.Modular and scalable architecture.

2.Easy integration with AI models and applications.

3.Provides reliable and secure access to AI services through standardized APIs.

# Task 3: Containerize AI Application

Containerization is the process of packaging an application along with its dependencies, libraries, and configuration files into a lightweight, portable unit called a container. Containers ensure that the application runs consistently across different environments such as development, testing, and production.

For AI applications, Docker is commonly used to containerize services like FastAPI-based APIs, machine learning models, and RAG systems.

Steps to Containerize an AI Application:

1.Create the AI application (e.g., FastAPI service).

2.Define dependencies in requirements.txt.

3.Create a Dockerfile to specify the runtime environment.

4.Build a Docker image.

5.Run the application inside a Docker container.

FastAPI Service:

```python
fastapi_code = '''import asyncio, logging, math, time, uuid, nest_asyncio
from datetime import datetime
from typing import List, Optional
from fastapi import FastAPI, Request
from fastapi.middleware.cors import CORSMiddleware
from fastapi.responses import JSONResponse
from pydantic import BaseModel, Field, field_validator, model_validator
import uvicorn

nest_asyncio.apply()
logging.basicConfig(level=logging.INFO, format="%(asctime)s | %(levelname)-8s | %(message)s")
logger = logging.getLogger("fastapi_service")

app = FastAPI(title="FastAPI Service", version="1.0.0")
app.add_middleware(CORSMiddleware, allow_origins=["*"], allow_methods=["*"], allow_headers=["*"])

class ChatMessage(BaseModel):
    role: str
    content: str = Field(..., min_length=1, max_length=32000)

    @field_validator("role")
    @classmethod
    def valid_role(cls, v):
        if v not in ("user", "assistant", "system"):
            raise ValueError("role must be user, assistant, or system")
        return v

class ChatRequest(BaseModel):
    messages: List[ChatMessage] = Field(..., min_length=1)
    model: Optional[str] = "gpt-4o"
    temperature: Optional[float] = Field(0.7, ge=0.0, le=2.0)
    max_tokens: Optional[int] = Field(1024, ge=1, le=8192)

    @model_validator(mode="after")
    def needs_user_msg(self):
        if not any(m.role == "user" for m in self.messages):
            raise ValueError("At least one message must have role=user")
        return self

class ChatResponse(BaseModel):
    id: str
    model: str
    created: int
    message: ChatMessage
    usage: dict
    latency_ms: float

@app.middleware("http")
async def log_requests(request: Request, call_next):
    rid = str(uuid.uuid4())[:8]
    logger.info("-> %s %s [%s]", request.method, request.url.path, rid)
    t0 = time.perf_counter()
    response = await call_next(request)
    logger.info("<- %.0fms [%s]", (time.perf_counter()-t0)*1000, rid)
    return response

@app.exception_handler(Exception)
async def global_err(request: Request, exc: Exception):
    logger.error("Error: %s", exc)
    return JSONResponse(status_code=500, content={"error": str(exc)})

@app.get("/health")
async def health():
    return {"status": "healthy", "service": "fastapi_service", "timestamp": datetime.utcnow().isoformat()+"Z"}

@app.post("/chat", response_model=ChatResponse)
async def chat(req: ChatRequest):
    t0 = time.perf_counter()
    await asyncio.sleep(0.05)
    last = next((m.content for m in reversed(req.messages) if m.role == "user"), "")
    replies = {
        "hello": "Hello! I am your FastAPI AI assistant.",
        "rag": "RAG combines vector search with LLM generation for grounded answers.",
        "agent": "AI agents autonomously plan and execute multi-step tasks using tools.",
    }
    reply = next((v for k, v in replies.items() if k in last.lower()), "FastAPI service response processed.")
    ms = (time.perf_counter()-t0)*1000
    pt = sum(len(m.content.split()) for m in req.messages)
    ct = len(reply.split())
    return ChatResponse(
        id=f"chat-{uuid.uuid4()}", model=req.model, created=int(time.time()),
        message=ChatMessage(role="assistant", content=reply),
        usage={"prompt_tokens": pt, "completion_tokens": ct, "total_tokens": pt+ct},
        latency_ms=round(ms, 2),
    )
'''

with open("ai_app/fastapi_service/main.py", "w") as f:
    f.write(fastapi_code)

print("✅ FastAPI service written")
```

Agent Service:

```python
agent_code = '''import asyncio, logging, time, uuid, nest_asyncio
from datetime import datetime
from typing import List, Optional, Dict, Any
from fastapi import FastAPI, Request
from fastapi.middleware.cors import CORSMiddleware
from fastapi.responses import JSONResponse
from pydantic import BaseModel, Field, field_validator
import uvicorn

nest_asyncio.apply()
logging.basicConfig(level=logging.INFO, format="%(asctime)s | %(levelname)-8s | %(message)s")
logger = logging.getLogger("agent_service")

app = FastAPI(title="AI Agent Service", version="1.0.0")
app.add_middleware(CORSMiddleware, allow_origins=["*"], allow_methods=["*"], allow_headers=["*"])

TOOLS = {
    "web_search":    {"description": "Search the web for current information", "params": ["query"]},
    "calculator":    {"description": "Perform mathematical calculations",       "params": ["expression"]},
    "file_reader":   {"description": "Read and analyze file contents",          "params": ["file_path"]},
    "code_executor": {"description": "Execute Python code safely",              "params": ["code"]},
    "summarizer":    {"description": "Summarize long documents",                "params": ["text", "max_length"]},
}

class AgentTask(BaseModel):
    task: str = Field(..., min_length=1, max_length=2000)
    tools: Optional[List[str]] = None
    max_steps: Optional[int] = Field(5, ge=1, le=20)
    context: Optional[Dict[str, Any]] = {}

    @field_validator("task")
    @classmethod
    def task_not_blank(cls, v):
        if not v.strip():
            raise ValueError("task must not be blank")
        return v

    @field_validator("tools", mode="before")
    @classmethod
    def validate_tools(cls, v):
        if v:
            invalid = [t for t in v if t not in TOOLS]
            if invalid:
                raise ValueError(f"Unknown tools: {invalid}. Available: {list(TOOLS.keys())}")
        return v

class AgentStep(BaseModel):
    step: int
    action: str
    tool: Optional[str]
    input: Optional[str]
    output: str
    latency_ms: float

class AgentResponse(BaseModel):
    id: str
    task: str
    steps: List[AgentStep]
    final_answer: str
    tools_used: List[str]
    total_steps: int
    total_latency_ms: float

async def mock_tool_call(tool: str, task: str) -> str:
    await asyncio.sleep(0.03)
    responses = {
        "web_search":    f"Found 5 results for: {task[:40]}. Top result: Comprehensive guide available.",
        "calculator":    "Calculation complete. Result: 42.0",
        "file_reader":   "File analyzed. Contains 1,200 words across 3 sections.",
        "code_executor": "Code executed successfully. Output: [1, 4, 9, 16, 25]",
        "summarizer":    f"Summary: The task '{task[:30]}...' involves multiple components requiring careful analysis.",
    }
    return responses.get(tool, "Tool executed successfully.")

@app.middleware("http")
async def log_requests(request: Request, call_next):
    rid = str(uuid.uuid4())[:8]
    logger.info("-> %s %s [%s]", request.method, request.url.path, rid)
    t0 = time.perf_counter()
    response = await call_next(request)
    logger.info("<- %.0fms [%s]", (time.perf_counter()-t0)*1000, rid)
    return response

@app.exception_handler(Exception)
async def global_err(request: Request, exc: Exception):
    logger.error("Error: %s", exc)
    return JSONResponse(status_code=500, content={"error": str(exc)})

@app.get("/health")
async def health():
    return {"status": "healthy", "service": "agent_service", "timestamp": datetime.utcnow().isoformat()+"Z",
            "available_tools": list(TOOLS.keys())}

@app.get("/tools")
async def list_tools():
    return {"tools": TOOLS, "total": len(TOOLS)}

@app.post("/agent/run", response_model=AgentResponse)
async def run_agent(req: AgentTask):
    logger.info("Agent task | steps=%d tools=%s", req.max_steps, req.tools)
    t_total = time.perf_counter()
    active_tools = req.tools or list(TOOLS.keys())[:3]
    steps = []
    tools_used = []

    step_plan = [
        ("Analyzing task and planning approach", None,             None),
        ("Gathering relevant information",       "web_search",     req.task),
        ("Processing and computing results",     "calculator",     "Processing task metrics"),
        ("Synthesizing findings",                "summarizer",     req.task),
        ("Generating final response",            None,             None),
    ]

    for i, (action, tool, inp) in enumerate(step_plan[:req.max_steps], 1):
        t0 = time.perf_counter()
        if tool and tool in active_tools:
            output = await mock_tool_call(tool, req.task)
            tools_used.append(tool)
        else:
            await asyncio.sleep(0.02)
            output = f"Step {i} completed: {action}"
        ms = (time.perf_counter()-t0)*1000
        steps.append(AgentStep(step=i, action=action, tool=tool, input=inp, output=output, latency_ms=round(ms,2)))
        logger.info("Step %d/%d | tool=%s | %.0fms", i, req.max_steps, tool or "none", ms)

    final = f"Task completed: '{req.task[:60]}'. Executed {len(steps)} steps using {len(tools_used)} tools. All objectives achieved."
    total_ms = (time.perf_counter()-t_total)*1000
    logger.info("Agent done | steps=%d tools_used=%d latency=%.1fms", len(steps), len(tools_used), total_ms)
    return AgentResponse(
        id=f"agent-{uuid.uuid4()}", task=req.task, steps=steps,
        final_answer=final, tools_used=list(set(tools_used)),
        total_steps=len(steps), total_latency_ms=round(total_ms, 2),
    )
'''

with open("ai_app/agent_service/main.py", "w") as f:
    f.write(agent_code)

print("✅ Agent service written")
```

RAG Service:

```python
rag_code = '''import asyncio, logging, math, time, uuid, nest_asyncio
from datetime import datetime
from typing import List, Optional, Dict
from fastapi import FastAPI, Request
from fastapi.middleware.cors import CORSMiddleware
from fastapi.responses import JSONResponse
from pydantic import BaseModel, Field, field_validator
import uvicorn

nest_asyncio.apply()
logging.basicConfig(level=logging.INFO, format="%(asctime)s | %(levelname)-8s | %(message)s")
logger = logging.getLogger("rag_service")

app = FastAPI(title="RAG Service", version="1.0.0")
app.add_middleware(CORSMiddleware, allow_origins=["*"], allow_methods=["*"], allow_headers=["*"])

VECTOR_STORE: Dict[str, dict] = {}
DOCUMENTS = [
    {"id": "d001", "content": "FastAPI is a modern Python web framework for building APIs with automatic validation.", "source": "docs/fastapi.md", "tags": ["python","api"]},
    {"id": "d002", "content": "Docker containerizes applications into portable, isolated environments called containers.", "source": "docs/docker.md", "tags": ["docker","devops"]},
    {"id": "d003", "content": "RAG retrieval-augmented generation combines vector search with LLM generation.", "source": "docs/rag.md", "tags": ["rag","llm","ai"]},
    {"id": "d004", "content": "Redis is an in-memory data store used for caching, sessions, and pub/sub messaging.", "source": "docs/redis.md", "tags": ["redis","cache"]},
    {"id": "d005", "content": "PostgreSQL is a powerful open-source relational database with ACID compliance.", "source": "docs/postgres.md", "tags": ["postgresql","database"]},
    {"id": "d006", "content": "Vector databases store embeddings and support similarity search at scale.", "source": "docs/vectordb.md", "tags": ["vector-db","embeddings"]},
    {"id": "d007", "content": "Kubernetes orchestrates containerized workloads across clusters with autoscaling.", "source": "docs/k8s.md", "tags": ["kubernetes","devops"]},
    {"id": "d008", "content": "Semantic caching serves similar queries from cache, reducing LLM API costs by up to 68%.", "source": "docs/caching.md", "tags": ["caching","cost"]},
]

def embed(text: str, dims: int = 16) -> List[float]:
    vec = [sum(ord(c)*(i+1) for c in text[:50]) % 1000 / 1000.0 for i in range(dims)]
    norm = math.sqrt(sum(x*x for x in vec)) or 1.0
    return [round(x/norm, 6) for x in vec]

def cosine(a, b):
    dot = sum(x*y for x,y in zip(a,b))
    na = math.sqrt(sum(x*x for x in a))
    nb = math.sqrt(sum(x*x for x in b))
    return dot/(na*nb) if na and nb else 0.0

for doc in DOCUMENTS:
    VECTOR_STORE[doc["id"]] = {**doc, "embedding": embed(doc["content"])}

class IngestRequest(BaseModel):
    documents: List[Dict] = Field(..., min_length=1)

    @field_validator("documents", mode="before")
    @classmethod
    def validate_docs(cls, v):
        for i, d in enumerate(v):
            if "content" not in d or not str(d["content"]).strip():
                raise ValueError(f"Document {i} must have non-empty content")
        return v

class RAGRequest(BaseModel):
    query: str = Field(..., min_length=1, max_length=1000)
    top_k: Optional[int] = Field(3, ge=1, le=10)
    threshold: Optional[float] = Field(0.0, ge=0.0, le=1.0)
    generate_answer: Optional[bool] = True

    @field_validator("query")
    @classmethod
    def not_blank(cls, v):
        if not v.strip():
            raise ValueError("query must not be blank")
        return v

class RetrievedChunk(BaseModel):
    id: str
    content: str
    score: float
    source: str
    tags: List[str]

class RAGResponse(BaseModel):
    id: str
    query: str
    retrieved_chunks: List[RetrievedChunk]
    generated_answer: Optional[str]
    total_chunks: int
    retrieval_ms: float
    generation_ms: float
    total_ms: float

@app.middleware("http")
async def log_requests(request: Request, call_next):
    rid = str(uuid.uuid4())[:8]
    logger.info("-> %s %s [%s]", request.method, request.url.path, rid)
    t0 = time.perf_counter()
    response = await call_next(request)
    logger.info("<- %.0fms [%s]", (time.perf_counter()-t0)*1000, rid)
    return response

@app.exception_handler(Exception)
async def global_err(request: Request, exc: Exception):
    logger.error("Error: %s", exc)
    return JSONResponse(status_code=500, content={"error": str(exc)})

@app.get("/health")
async def health():
    return {"status": "healthy", "service": "rag_service",
            "documents_indexed": len(VECTOR_STORE), "timestamp": datetime.utcnow().isoformat()+"Z"}

@app.post("/ingest")
async def ingest(req: IngestRequest):
    ingested = []
    for doc in req.documents:
        did = doc.get("id", f"doc-{uuid.uuid4()}")
        VECTOR_STORE[did] = {
            "id": did, "content": doc["content"],
            "source": doc.get("source", "user_upload"),
            "tags": doc.get("tags", []),
            "embedding": embed(doc["content"]),
        }
        ingested.append(did)
    logger.info("Ingested %d documents. Total: %d", len(ingested), len(VECTOR_STORE))
    return {"ingested": len(ingested), "ids": ingested, "total_indexed": len(VECTOR_STORE)}

@app.post("/rag/query", response_model=RAGResponse)
async def rag_query(req: RAGRequest):
    logger.info("RAG query | query='%s' top_k=%d generate=%s", req.query[:60], req.top_k, req.generate_answer)
    t_total = time.perf_counter()

    t_ret = time.perf_counter()
    await asyncio.sleep(0.03)
    qv = embed(req.query)
    scored = sorted([(cosine(qv, v["embedding"]), v) for v in VECTOR_STORE.values()],
                    key=lambda x: x[0], reverse=True)
    chunks = [
        RetrievedChunk(id=d["id"], content=d["content"], score=round(s,4),
                       source=d["source"], tags=d.get("tags",[]))
        for s, d in scored[:req.top_k] if s >= req.threshold
    ]
    ret_ms = (time.perf_counter()-t_ret)*1000

    answer = None
    gen_ms = 0.0
    if req.generate_answer and chunks:
        t_gen = time.perf_counter()
        await asyncio.sleep(0.05)
        context = " | ".join(c.content[:80] for c in chunks[:2])
        answer = (f"Based on {len(chunks)} retrieved documents: {context[:200]}... "
                  f"This answer is grounded in your knowledge base with confidence {chunks[0].score:.2f}.")
        gen_ms = (time.perf_counter()-t_gen)*1000

    total_ms = (time.perf_counter()-t_total)*1000
    logger.info("RAG done | chunks=%d ret=%.0fms gen=%.0fms total=%.0fms",
                len(chunks), ret_ms, gen_ms, total_ms)
    return RAGResponse(
        id=f"rag-{uuid.uuid4()}", query=req.query,
        retrieved_chunks=chunks, generated_answer=answer,
        total_chunks=len(chunks), retrieval_ms=round(ret_ms,2),
        generation_ms=round(gen_ms,2), total_ms=round(total_ms,2),
    )
'''

with open("ai_app/rag_service/main.py", "w") as f:
    f.write(rag_code)

print("✅ RAG service written")
```

# Task 6: LLM Monitoring

LLM Monitoring is the process of tracking and analyzing the performance, usage, and reliability of Large Language Models (LLMs) in production environments.

Metrics Tracked

1.Request Count → Total number of API requests.

2.Response Time → Average latency for generating responses.

3.Token Usage → Total tokens consumed.

4.Prompt Length → Length of input prompts.

5.Completion Length → Length of generated responses.

6.Error Rate → Percentage of failed requests.

Create llm_monitor.py:

```python
llm_monitor_code = '''
import json
from datetime import datetime

class LLMMonitor:

    def __init__(self):
        self.metrics = {
            "requests": 1500,
            "avg_latency": 1200,
            "token_usage": 450000,
            "prompt_length": 120,
            "completion_length": 250,
            "error_rate": 0.02,
            "timestamp": str(datetime.now())
        }

    def generate_report(self):
        return self.metrics

    def save_metrics(self):
        with open("llm_metrics.json", "w") as f:
            json.dump(self.metrics, f, indent=4)

        print("Metrics saved to llm_metrics.json")


if __name__ == "__main__":

    monitor = LLMMonitor()

    metrics = monitor.generate_report()

    print("\\nLLM Monitoring Metrics\\n")
    print(json.dumps(metrics, indent=4))

    monitor.save_metrics()
'''

with open("llm_monitor.py", "w") as f:
    f.write(llm_monitor_code)

print("llm_monitor.py created successfully")
```

# Task 9: Incident Detection & Alerting

An Incident Detection & Alerting System continuously monitors AI applications and infrastructure to identify abnormal conditions and notify administrators before they impact users. It helps maintain system reliability, availability, and performance.

Alerts Implemented:

1. High Error Rate

Triggered when API or application errors exceed a predefined threshold.
Indicates potential bugs, failures, or service instability.

2. High Cost

Triggered when operational costs exceed budget limits.
Helps control AI infrastructure and model usage expenses.

3. Service Downtime

Triggered when services become unavailable.
Ensures quick response to outages.

4. Slow Response Times

Triggered when latency exceeds acceptable limits.
Helps maintain a good user experience.

5. Failed Retrievals

Triggered when document retrieval or search operations fail.
Important for RAG and retrieval-based AI systems.

Alert Channels:

1.Email:

Sends notifications directly to administrators.
Suitable for formal incident reporting.

2.Slack:

Provides real-time team notifications.
Supports rapid collaboration and response.

3.Microsoft Teams:

Delivers alerts to operations and support teams.
Enables enterprise communication workflows.

Incident Detection & Alerting:

```python
alerting_code = '''
from datetime import datetime

class AlertingSystem:

    def __init__(self):

        self.metrics = {
            "error_rate": 0.08,
            "daily_cost": 150,
            "service_status": "DOWN",
            "latency_ms": 3500,
            "retrieval_success": False
        }

        self.alerts = []

    def detect_incidents(self):

        if self.metrics["error_rate"] > 0.05:
            self.alerts.append("HIGH ERROR RATE")

        if self.metrics["daily_cost"] > 100:
            self.alerts.append("HIGH COST")

        if self.metrics["service_status"] != "UP":
            self.alerts.append("SERVICE DOWNTIME")

        if self.metrics["latency_ms"] > 2000:
            self.alerts.append("SLOW RESPONSE TIMES")

        if not self.metrics["retrieval_success"]:
            self.alerts.append("FAILED RETRIEVALS")

    def send_email_alert(self):
        print("Email Alert Sent")

    def send_slack_alert(self):
        print("Slack Alert Sent")

    def send_teams_alert(self):
        print("Teams Alert Sent")

    def generate_report(self):

        report = f"""
# Incident Report

Generated: {datetime.now()}

## Detected Incidents

"""

        for alert in self.alerts:
            report += f"- {alert}\\n"

        report += """

## Alert Channels

- Email
- Slack
- Microsoft Teams

## Recommended Actions

1. Investigate API failures.
2. Review infrastructure health.
3. Optimize AI model costs.
4. Improve response latency.
5. Verify retrieval pipeline.

## Status

Incident Detection Completed.
"""

        with open("incident_report.md", "w") as f:
            f.write(report)

        print("incident_report.md generated successfully")

if __name__ == "__main__":

    system = AlertingSystem()

    system.detect_incidents()

    print("\\nDetected Alerts:")
    for alert in system.alerts:
        print("-", alert)

    system.send_email_alert()
    system.send_slack_alert()
    system.send_teams_alert()

    system.generate_report()
'''

with open("alerting_system.py", "w") as f:
    f.write(alerting_code)

print("alerting_system.py created successfully")

!python alerting_system.py
```
