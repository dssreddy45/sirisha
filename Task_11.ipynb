# Task 2: Build First Tool Calling Agent

Build First Tool Calling Agent:

A Tool Calling Agent is an AI agent that can use external tools, APIs, databases, or services to perform tasks beyond generating text.

Steps to Build:

Define the user input
Create available tools/functions
Connect the AI model to those tools
Enable function calling logic
Execute tools and collect results
Return the final response to the user

Example:

User → “Find today’s weather and send summary”

AI Agent → Calls weather API → Receives data → Generates response.

````python
!pip install groq -q

import json
from datetime import datetime
from groq import Groq

client = Groq(api_key="API KEY")  # paste your Groq key
MODEL  = "llama-3.3-70b-versatile"

# ============================================================
# TOOLS DEFINITION
# ============================================================

def calculator(expression: str) -> str:
    try:
        result = eval(expression)
        return str(result)
    except Exception as e:
        return f"Error: {e}"

def get_current_date() -> str:
    return datetime.now().strftime("%A, %B %d, %Y %I:%M %p")

def text_summary(text: str) -> str:
    response = client.chat.completions.create(
        model=MODEL,
        messages=[{"role": "user", "content": f"Summarize this in 2-3 sentences:\n\n{text}"}],
        max_tokens=200
    )
    return response.choices[0].message.content.strip()

# ============================================================
# TOOLS SCHEMA (for LLM)
# ============================================================

tools = [
    {
        "type": "function",
        "function": {
            "name": "calculator",
            "description": "Evaluate a math expression like 250 * 45 or 100 + 200",
            "parameters": {
                "type": "object",
                "properties": {
                    "expression": {"type": "string", "description": "Math expression to evaluate"}
                },
                "required": ["expression"]
            }
        }
    },
    {
        "type": "function",
        "function": {
            "name": "get_current_date",
            "description": "Get the current system date and time",
            "parameters": {
                "type": "object",
                "properties": {}
            }
        }
    },
    {
        "type": "function",
        "function": {
            "name": "text_summary",
            "description": "Summarize a long piece of text into 2-3 sentences",
            "parameters": {
                "type": "object",
                "properties": {
                    "text": {"type": "string", "description": "Long text to summarize"}
                },
                "required": ["text"]
            }
        }
    }
]

# ============================================================
# TOOL EXECUTOR
# ============================================================

def execute_tool(tool_name, tool_args):
    print(f"\n🔧 Calling tool: {tool_name}")
    print(f"   Args: {tool_args}")
    if tool_name == "calculator":
        return calculator(tool_args["expression"])
    elif tool_name == "get_current_date":
        return get_current_date()
    elif tool_name == "text_summary":
        return text_summary(tool_args["text"])
    else:
        return "Unknown tool"

# ============================================================
# AGENT
# ============================================================

def agent(user_input):
    print(f"\n{'='*55}")
    print(f"👤 User: {user_input}")
    print(f"{'='*55}")

    messages = [{"role": "user", "content": user_input}]

    # Step 1: Ask LLM which tool to call
    response = client.chat.completions.create(
        model=MODEL,
        messages=messages,
        tools=tools,
        tool_choice="auto"
    )

    msg = response.choices[0].message

    # Step 2: If no tool needed, return direct answer
    if not msg.tool_calls:
        print(f"\n💬 Answer: {msg.content}")
        return msg.content

    # Step 3: Execute the tool
    tool_call   = msg.tool_calls[0]
    tool_name   = tool_call.function.name
    tool_args   = json.loads(tool_call.function.arguments)
    tool_result = execute_tool(tool_name, tool_args)

    print(f"   Result: {tool_result}")

    # Step 4: Send result back to LLM for final answer
    messages.append({"role": "assistant", "content": None, "tool_calls": msg.tool_calls})
    messages.append({"role": "tool", "tool_call_id": tool_call.id, "content": tool_result})

    final = client.chat.completions.create(
        model=MODEL,
        messages=messages
    )

    answer = final.choices[0].message.content.strip()
    print(f"\n💬 Answer: {answer}")
    return answer

# ============================================================
# DEMO
# ============================================================

print("\n" + "🤖"*20)
print("     TOOL CALLING AGENT DEMO")
print("🤖"*20)

# Test 1: Calculator
agent("What is 250 * 45?")

# Test 2: Current Date
agent("What is today's date and time?")

# Test 3: Text Summary
agent("""Summarize this: Artificial Intelligence is transforming every industry.
From healthcare to finance, AI systems are automating complex tasks,
improving accuracy, reducing costs, and enabling new capabilities
that were previously impossible. Machine learning models trained on
vast datasets can now diagnose diseases, predict stock movements,
generate creative content, write code, and drive vehicles autonomously.
The pace of AI development has accelerated dramatically in recent years,
with large language models like GPT and Claude demonstrating remarkable
reasoning and language understanding capabilities.""")

# Save to file
code = open(__file__).read() if "__file__" in dir() else "# tool_calling_agent.py\n# Run this code in Colab"
with open("tool_calling_agent.py", "w") as f:
    f.write(code)
print("\n Saved: tool_calling_agent.py")

from google.colab import files
files.download("tool_calling_agent.py")
````

# Task 3: Create Function Calling Framework

Function Calling Framework:

A Function Calling Framework is a structure that allows an AI model to identify tasks, call external functions/tools, process results, and return responses automatically.

Core Components:

User Input – Receives the request
Function Definitions – Available tools and operations
Decision Logic – Selects the correct function
Execution Layer – Runs API/tool calls
Response Generator – Converts outputs into user-friendly responses

Example:
User → “Check weather in Chennai”

AI → Detects weather function → Calls weather API → Receives data → Returns weather report.

````python
!pip install groq -q

import json
from groq import Groq

client = Groq(api_key="API KEY")  # paste your Groq key
MODEL  = "llama-3.3-70b-versatile"

# ============================================================
# MOCK DATABASES & APIs
# ============================================================

# Mock Weather API
WEATHER_DB = {
    "hyderabad": {"temp": 34, "condition": "Sunny",  "humidity": "60%", "wind": "12 km/h"},
    "mumbai":    {"temp": 30, "condition": "Cloudy", "humidity": "80%", "wind": "18 km/h"},
    "delhi":     {"temp": 38, "condition": "Hot",    "humidity": "45%", "wind": "8 km/h"},
    "bangalore": {"temp": 26, "condition": "Rainy",  "humidity": "75%", "wind": "15 km/h"},
}

# Mock Currency API
CURRENCY_DB = {
    ("USD", "INR"): 83.5,
    ("EUR", "INR"): 91.2,
    ("GBP", "INR"): 106.3,
    ("INR", "USD"): 0.012,
    ("INR", "EUR"): 0.011,
    ("USD", "EUR"): 0.92,
    ("EUR", "USD"): 1.09,
}

# Mock Employee Database
EMPLOYEE_DB = {
    "EMP001": {"id": "EMP001", "name": "Ajay",   "department": "Engineering", "role": "Senior Developer",  "email": "ajay@techcorp.com",   "salary": "₹12,00,000"},
    "EMP002": {"id": "EMP002", "name": "Priya",  "department": "HR",          "role": "HR Manager",        "email": "priya@techcorp.com",  "salary": "₹9,00,000"},
    "EMP003": {"id": "EMP003", "name": "Rahul",  "department": "Finance",     "role": "Financial Analyst", "email": "rahul@techcorp.com",  "salary": "₹10,00,000"},
    "EMP004": {"id": "EMP004", "name": "Sneha",  "department": "Marketing",   "role": "Marketing Lead",    "email": "sneha@techcorp.com",  "salary": "₹8,50,000"},
    "EMP005": {"id": "EMP005", "name": "Vikram", "department": "Engineering", "role": "DevOps Engineer",   "email": "vikram@techcorp.com", "salary": "₹11,00,000"},
}

# ============================================================
# TOOL FUNCTIONS
# ============================================================

def get_weather(city: str) -> dict:
    city = city.lower().strip()
    if city in WEATHER_DB:
        data = WEATHER_DB[city]
        return {
            "city":      city.title(),
            "temp":      f"{data['temp']}°C",
            "condition": data["condition"],
            "humidity":  data["humidity"],
            "wind":      data["wind"]
        }
    return {"error": f"Weather data not found for '{city}'"}

def convert_currency(amount: float, from_currency: str, to_currency: str) -> dict:
    from_currency = from_currency.upper().strip()
    to_currency   = to_currency.upper().strip()
    key = (from_currency, to_currency)
    if key in CURRENCY_DB:
        rate   = CURRENCY_DB[key]
        result = round(amount * rate, 2)
        return {
            "from":   f"{amount} {from_currency}",
            "to":     f"{result} {to_currency}",
            "rate":   rate,
            "result": result
        }
    return {"error": f"Conversion rate not found for {from_currency} to {to_currency}"}

def get_employee(employee_id: str) -> dict:
    employee_id = employee_id.upper().strip()
    if employee_id in EMPLOYEE_DB:
        return EMPLOYEE_DB[employee_id]
    return {"error": f"Employee '{employee_id}' not found in database"}

# ============================================================
# TOOLS SCHEMA
# ============================================================

tools = [
    {
        "type": "function",
        "function": {
            "name": "get_weather",
            "description": "Get current weather for a city using mock API",
            "parameters": {
                "type": "object",
                "properties": {
                    "city": {"type": "string", "description": "City name e.g. Hyderabad, Mumbai"}
                },
                "required": ["city"]
            }
        }
    },
    {
        "type": "function",
        "function": {
            "name": "convert_currency",
            "description": "Convert currency from one type to another using mock API",
            "parameters": {
                "type": "object",
                "properties": {
                    "amount":        {"type": "number", "description": "Amount to convert"},
                    "from_currency": {"type": "string", "description": "Source currency e.g. USD, EUR, INR"},
                    "to_currency":   {"type": "string", "description": "Target currency e.g. USD, EUR, INR"}
                },
                "required": ["amount", "from_currency", "to_currency"]
            }
        }
    },
    {
        "type": "function",
        "function": {
            "name": "get_employee",
            "description": "Look up employee information from mock database using employee ID",
            "parameters": {
                "type": "object",
                "properties": {
                    "employee_id": {"type": "string", "description": "Employee ID e.g. EMP001, EMP002"}
                },
                "required": ["employee_id"]
            }
        }
    }
]

# ============================================================
# TOOL EXECUTOR
# ============================================================

def execute_tool(tool_name, tool_args):
    print(f"\n🔧 Tool Called : {tool_name}")
    print(f"   Arguments   : {json.dumps(tool_args, indent=2)}")
    if tool_name == "get_weather":
        result = get_weather(tool_args["city"])
    elif tool_name == "convert_currency":
        result = convert_currency(tool_args["amount"], tool_args["from_currency"], tool_args["to_currency"])
    elif tool_name == "get_employee":
        result = get_employee(tool_args["employee_id"])
    else:
        result = {"error": "Unknown tool"}
    print(f"   Result      : {json.dumps(result, indent=2)}")
    return json.dumps(result)

# ============================================================
# FUNCTION CALLING FRAMEWORK
# ============================================================

def run_agent(user_input):
    print(f"\n{'='*60}")
    print(f"👤 User : {user_input}")
    print(f"{'='*60}")

    messages = [{"role": "user", "content": user_input}]

    # Step 1: LLM decides which tool to call
    response = client.chat.completions.create(
        model=MODEL,
        messages=messages,
        tools=tools,
        tool_choice="auto"
    )

    msg = response.choices[0].message

    # Step 2: No tool needed
    if not msg.tool_calls:
        print(f"\n💬 Answer: {msg.content}")
        return msg.content

    # Step 3: Execute all tool calls
    messages.append({
        "role":       "assistant",
        "content":    None,
        "tool_calls": msg.tool_calls
    })

    for tool_call in msg.tool_calls:
        tool_name   = tool_call.function.name
        tool_args   = json.loads(tool_call.function.arguments)
        tool_result = execute_tool(tool_name, tool_args)
        messages.append({
            "role":         "tool",
            "tool_call_id": tool_call.id,
            "content":      tool_result
        })

    # Step 4: Final answer from LLM
    final = client.chat.completions.create(
        model=MODEL,
        messages=messages
    )

    answer = final.choices[0].message.content.strip()
    print(f"\n💬 Answer: {answer}")
    return answer

# ============================================================
# DEMO
# ============================================================

print("\n" + "🚀"*20)
print("   FUNCTION CALLING FRAMEWORK DEMO")
print("🚀"*20)

# Test 1: Employee Lookup
run_agent("Show employee information for EMP001")

# Test 2: Weather
run_agent("What is the weather in Hyderabad?")

# Test 3: Currency
run_agent("Convert 100 USD to INR")

# Test 4: All employees
run_agent("Get details of employee EMP003 and EMP005")

# Test 5: Combined
run_agent("What is the weather in Mumbai and convert 500 EUR to INR?")

# ============================================================
# SAVE FILE
# ============================================================

file_content = '''# function_calling.py
# Run this file in Google Colab

# Install: pip install groq

# This framework includes:
# 1. Weather Tool (Mock API)
# 2. Currency Tool (Mock API)
# 3. Employee Lookup Tool (Mock Database)
'''

with open("function_calling.py", "w") as f:
    f.write(file_content)

print("\n✅ Saved: function_calling.py")

from google.colab import files
files.download("function_calling.py")
````

# Task 4: Multi-Step Agent Workflow

Multi-Step Agent Workflow:

A Multi-Step Agent Workflow is a process where an AI agent breaks a complex task into multiple smaller steps, executes them sequentially, and combines the results to achieve a final goal.

Key Features:

Task decomposition
Sequential decision-making
Tool and API integration
Memory and context handling
Automated execution

Example:
User → “Plan my trip”

Agent → Searches flights → Books hotel → Creates itinerary → Sends final travel plan.

````python
!pip install groq -q

from groq import Groq

client = Groq(api_key="API KEY")  # paste your Groq key
MODEL  = "llama-3.3-70b-versatile"

# ============================================================
# HELPER: Call LLM
# ============================================================

def llm(prompt, max_tokens=1000):
    response = client.chat.completions.create(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}],
        max_tokens=max_tokens
    )
    return response.choices[0].message.content.strip()

# ============================================================
# STEP FUNCTIONS
# ============================================================

def step1_requirements(project):
    print("\n📋 Step 1: Generating Requirements...")
    prompt = f"""You are a software architect.
Generate detailed functional and non-functional requirements for: {project}

Format:
FUNCTIONAL REQUIREMENTS:
1. ...
2. ...

NON-FUNCTIONAL REQUIREMENTS:
1. ...
2. ...

Keep it concise — 8-10 points each."""
    result = llm(prompt)
    print(result)
    return result

def step2_user_stories(project, requirements):
    print("\n👤 Step 2: Generating User Stories...")
    prompt = f"""You are a product manager.
Based on these requirements for {project}:
{requirements}

Generate 8 user stories in this format:
As a [role], I want to [action] so that [benefit].

Include stories for: Rider, Driver, Admin."""
    result = llm(prompt)
    print(result)
    return result

def step3_database_schema(project, requirements):
    print("\n🗄️ Step 3: Generating Database Schema...")
    prompt = f"""You are a database architect.
Design a database schema for {project} based on:
{requirements}

For each table provide:
TABLE_NAME (
  field_name: datatype - description,
  ...
)

Include: Users, Drivers, Rides, Payments, Ratings tables."""
    result = llm(prompt)
    print(result)
    return result

def step4_api_list(project, requirements):
    print("\n🔌 Step 4: Generating API List...")
    prompt = f"""You are a backend engineer.
Design REST APIs for {project} based on:
{requirements}

Format each API as:
METHOD /endpoint - Description

Group by: Auth, User, Driver, Ride, Payment APIs.
Include 15-20 endpoints."""
    result = llm(prompt)
    print(result)
    return result

def step5_roadmap(project, requirements):
    print("\n🗺️ Step 5: Generating Development Roadmap...")
    prompt = f"""You are a project manager.
Create a development roadmap for {project} based on:
{requirements}

Format:
PHASE 1 - [Name] (Week X-X):
- Task 1
- Task 2

Include 4 phases: MVP, Core Features, Advanced Features, Launch."""
    result = llm(prompt)
    print(result)
    return result

# ============================================================
# MULTI-STEP AGENT
# ============================================================

def multi_step_agent(project):
    print("\n" + "🚀"*20)
    print(f"  PROJECT: {project}")
    print("🚀"*20)

    package = {"project": project}

    # Step 1
    print(f"\n{'='*60}")
    print("STEP 1/5 — REQUIREMENTS")
    print('='*60)
    package["requirements"] = step1_requirements(project)

    # Step 2
    print(f"\n{'='*60}")
    print("STEP 2/5 — USER STORIES")
    print('='*60)
    package["user_stories"] = step2_user_stories(project, package["requirements"])

    # Step 3
    print(f"\n{'='*60}")
    print("STEP 3/5 — DATABASE SCHEMA")
    print('='*60)
    package["database_schema"] = step3_database_schema(project, package["requirements"])

    # Step 4
    print(f"\n{'='*60}")
    print("STEP 4/5 — API LIST")
    print('='*60)
    package["api_list"] = step4_api_list(project, package["requirements"])

    # Step 5
    print(f"\n{'='*60}")
    print("STEP 5/5 — DEVELOPMENT ROADMAP")
    print('='*60)
    package["roadmap"] = step5_roadmap(project, package["requirements"])

    return package

# ============================================================
# SAVE RESULTS
# ============================================================

def save_package(package):
    content = f"""# Project Plan: {package['project']}
Generated by Multi-Step Agent

{'='*60}
## STEP 1: REQUIREMENTS
{'='*60}
{package['requirements']}

{'='*60}
## STEP 2: USER STORIES
{'='*60}
{package['user_stories']}

{'='*60}
## STEP 3: DATABASE SCHEMA
{'='*60}
{package['database_schema']}

{'='*60}
## STEP 4: API LIST
{'='*60}
{package['api_list']}

{'='*60}
## STEP 5: DEVELOPMENT ROADMAP
{'='*60}
{package['roadmap']}
"""
    with open("project_plan.md", "w") as f:
        f.write(content)

    with open("multi_step_agent.py", "w") as f:
        f.write("# multi_step_agent.py\n# Multi-Step Agent Workflow\n# Run in Google Colab\n")

    print("\n Saved: project_plan.md")
    print(" Saved: multi_step_agent.py")

# ============================================================
# RUN
# ============================================================

package = multi_step_agent("Create a project plan for a Ride Booking App")
save_package(package)

print("\n" + ""*20)
print("  COMPLETE PACKAGE GENERATED!")
print(""*20)
print(f"\n Package contains:")
for key in package:
    if key != "project":
        print(f"    {key.replace('_', ' ').title()}")

from google.colab import files
files.download("project_plan.md")
files.download("multi_step_agent.py")
````

````python

MODEL = "llama-3.3-70b-versatile"

SYSTEM_PROMPT = """You are an intelligent HR Agent. Help employees with:
1. HR Policy Questions (leave, WFH, conduct, performance)
2. Leave Balance Checks
3. Payroll FAQs (salary, payslips, tax, reimbursements, PF/Gratuity)
4. Onboarding Guidance for new joiners

Rules:
- Always use the available tools to fetch accurate data.
- Call ONE tool at a time. Wait for the result before calling another.
- ALWAYS cite the source in your response (e.g., Source: HR Policy Manual v3.2, Section 4).
- Be professional, clear, and empathetic.
- If info is unavailable, direct to hr@company.com."""

def run_hr_agent(user_query: str, api_key: str, verbose: bool = True) -> str:
    client = Groq(api_key=api_key)
    messages = [
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "user",   "content": user_query}
    ]

    if verbose:
        print(f"\n{'='*60}")
        print(f" HR AGENT  |  Model: {MODEL}")
        print(f"{'='*60}")
        print(f" Query: {user_query}")
        print(f"{'='*60}")

    max_iterations = 6  # prevent infinite loops

    for iteration in range(max_iterations):
        try:
            response = client.chat.completions.create(
                model=MODEL,
                messages=messages,
                tools=tools,
                tool_choice="auto",
                max_tokens=2000,
                parallel_tool_calls=False,   #  KEY FIX: one tool at a time
            )
        except Exception as e:
            # If tool call fails, retry without tools
            if verbose:
                print(f"  Tool call failed ({e}). Retrying without tools...")
            response = client.chat.completions.create(
                model=MODEL,
                messages=messages,
                max_tokens=2000,
            )
            final_answer = response.choices[0].message.content or ""
            if verbose:
                print(f"\n HR Agent Response (fallback):")
                print(f"{'-'*60}")
                print(final_answer)
                print(f"{'='*60}\n")
            return final_answer

        msg = response.choices[0].message

        # Safely build assistant message
        assistant_msg = {"role": "assistant", "content": msg.content or ""}
        if msg.tool_calls:
            assistant_msg["tool_calls"] = [
                {
                    "id": tc.id,
                    "type": "function",
                    "function": {
                        "name": tc.function.name,
                        "arguments": tc.function.arguments
                    }
                }
                for tc in msg.tool_calls
            ]
        messages.append(assistant_msg)

        # Execute tool calls if present
        if msg.tool_calls:
            for tc in msg.tool_calls:
                fn_name = tc.function.name

                # Safely parse arguments
                try:
                    fn_input = json.loads(tc.function.arguments)
                except json.JSONDecodeError:
                    fn_input = {}

                if verbose:
                    print(f"\n Tool: {fn_name}")
                    print(f"   Input: {json.dumps(fn_input, indent=2)}")

                result = execute_tool(fn_name, fn_input)

                if verbose:
                    preview = result[:200] + ("..." if len(result) > 200 else "")
                    print(f"    Result: {preview}")

                messages.append({
                    "role": "tool",
                    "tool_call_id": tc.id,
                    "content": result
                })
        else:
            # No tool calls = final answer
            final_answer = msg.content or ""
            if verbose:
                print(f"\n HR Agent Response:")
                print(f"{'-'*60}")
                print(final_answer)
                print(f"{'='*60}\n")
            return final_answer

    return "Max iterations reached. Please try again."
````

# Task 6: Agent Memory System

Agent Memory System

An Agent Memory System enables an AI agent to remember information across interactions, use past experiences, and improve decision-making.

Why Memory is Needed

Without memory:

User: My name is Sirisha.
Agent: Nice to meet you.

(New conversation)

User: What's my name?
Agent: I don't know.

With memory:

User: My name is Sirisha.
Agent: Nice to meet you.

(New conversation)

User: What's my name?
Agent: Your name is Sirisha.

Types of Memory

1. Short-Term Memory

2. Long-Term Memory

3. Episodic Memory

4. Semantic Memory

5. Procedural Memory

````python

import json
import os
import re
from datetime import datetime
from groq import Groq

# ============================================================
# 1. LONG-TERM MEMORY
# ============================================================

MEMORY_FILE = "long_term_memory.json"

def load_long_term_memory() -> dict:
    if os.path.exists(MEMORY_FILE):
        with open(MEMORY_FILE, "r") as f:
            return json.load(f)
    return {}

def save_long_term_memory(memory: dict):
    with open(MEMORY_FILE, "w") as f:
        json.dump(memory, f, indent=2)

def update_long_term_memory(employee_id: str, key: str, value):
    memory = load_long_term_memory()
    if employee_id not in memory:
        memory[employee_id] = {
            "employee_id": employee_id,
            "preferred_language": "English",
            "created_at": datetime.now().isoformat()
        }
    memory[employee_id][key] = value
    memory[employee_id]["updated_at"] = datetime.now().isoformat()
    save_long_term_memory(memory)
    return memory[employee_id]

def get_long_term_memory(employee_id: str) -> dict:
    memory = load_long_term_memory()
    return memory.get(employee_id, {})

# ============================================================
# 2. SHORT-TERM MEMORY
# ============================================================

class ShortTermMemory:
    def __init__(self, max_history: int = 20):
        self.conversation_history = []
        self.max_history = max_history
        self.session_facts = {}

    def add_message(self, role: str, content: str):
        self.conversation_history.append({
            "role": role,
            "content": content,
            "timestamp": datetime.now().isoformat()
        })
        if len(self.conversation_history) > self.max_history:
            self.conversation_history = self.conversation_history[-self.max_history:]

    def get_messages_for_api(self) -> list:
        return [
            {"role": msg["role"], "content": msg["content"]}
            for msg in self.conversation_history
        ]

    def add_session_fact(self, key: str, value):
        self.session_facts[key] = value

    def get_session_fact(self, key: str):
        return self.session_facts.get(key)

    def clear(self):
        self.conversation_history = []
        self.session_facts = {}

    def display(self):
        print("\n SHORT-TERM MEMORY")
        print("-" * 40)
        print(f"Messages in history: {len(self.conversation_history)}")
        for i, msg in enumerate(self.conversation_history):
            icon = "👤" if msg["role"] == "user" else "🤖"
            print(f"  [{i+1}] {icon} {msg['role'].upper()}: {msg['content'][:80]}{'...' if len(msg['content']) > 80 else ''}")
        if self.session_facts:
            print(f"\nSession Facts: {json.dumps(self.session_facts, indent=2)}")

# ============================================================
# 3. MEMORY MANAGER
# ============================================================

class MemoryManager:
    def __init__(self, employee_id: str = "GUEST"):
        self.employee_id = employee_id
        self.short_term = ShortTermMemory(max_history=20)
        self.long_term = get_long_term_memory(employee_id)

        if not self.long_term:
            self.long_term = {
                "employee_id": employee_id,
                "preferred_language": "English"
            }

    def remember(self, key: str, value, permanent: bool = False):
        self.short_term.add_session_fact(key, value)
        if permanent:
            self.long_term[key] = value
            update_long_term_memory(self.employee_id, key, value)

    def recall(self, key: str):
        val = self.short_term.get_session_fact(key)
        if val is not None:
            return val
        return self.long_term.get(key)

    def build_system_prompt(self) -> str:
        lt_context = json.dumps(self.long_term, indent=2) if self.long_term else "No long-term memory yet."
        return f"""You are a helpful AI Agent with memory capabilities.

 LONG-TERM MEMORY (persists across sessions):
{lt_context}

 SHORT-TERM MEMORY: You have the full conversation history above.

MEMORY RULES:
- If the user tells you their name, remember it and use it in responses.
- If asked recall questions like "what is my name?", answer from memory.
- Always be consistent — never forget what was said earlier.
- Greet returning users by name if you know it.

Be conversational, helpful, and show that you remember the user."""

    def display_all_memory(self):
        print("\n" + "="*50)
        print(" MEMORY MANAGER STATUS")
        print("="*50)
        print(f"\n LONG-TERM MEMORY (Employee: {self.employee_id})")
        print("-"*40)
        if self.long_term:
            for k, v in self.long_term.items():
                print(f"  {k}: {v}")
        else:
            print("  (empty)")
        self.short_term.display()
        print("="*50 + "\n")

# ============================================================
# 4. FACT EXTRACTOR
# ============================================================

def extract_and_store_facts(user_message: str, memory: MemoryManager):
    msg_lower = user_message.lower()

    # Detect name
    for trigger in ["my name is ", "i am ", "i'm ", "call me "]:
        if trigger in msg_lower:
            idx = msg_lower.index(trigger) + len(trigger)
            name = user_message[idx:].strip().split()[0].rstrip(".,!")
            if name:
                memory.remember("name", name, permanent=True)
                print(f"    Stored name: '{name}' → Long-Term Memory")

    # Detect language
    if "hindi" in msg_lower:
        memory.remember("preferred_language", "Hindi", permanent=True)
        print(f"    Stored language: 'Hindi' → Long-Term Memory")
    elif "telugu" in msg_lower:
        memory.remember("preferred_language", "Telugu", permanent=True)
        print(f"    Stored language: 'Telugu' → Long-Term Memory")
    elif "english" in msg_lower:
        memory.remember("preferred_language", "English", permanent=True)
        print(f"    Stored language: 'English' → Long-Term Memory")

    # Detect employee ID
    emp_match = re.search(r'\b(EMP\d{3,6}|E\d{3})\b', user_message, re.IGNORECASE)
    if emp_match:
        emp_id = emp_match.group().upper()
        memory.remember("employee_id", emp_id, permanent=True)
        print(f"    Stored employee_id: '{emp_id}' → Long-Term Memory")

# ============================================================
# 5. AGENT TURN
# ============================================================

MODEL = "llama-3.3-70b-versatile"

def run_agent_turn(user_message: str, memory: MemoryManager, client: Groq) -> str:
    extract_and_store_facts(user_message, memory)
    memory.short_term.add_message("user", user_message)

    messages = [
        {"role": "system", "content": memory.build_system_prompt()}
    ] + memory.short_term.get_messages_for_api()

    response = client.chat.completions.create(
        model=MODEL,
        messages=messages,
        max_tokens=1000,
        temperature=0.7,
    )

    agent_reply = response.choices[0].message.content or ""
    memory.short_term.add_message("assistant", agent_reply)
    return agent_reply

# ============================================================
# 6. DEMO
# ============================================================

def run_memory_demo(api_key: str):
    print("\n" + " "*20)
    print("       AGENT MEMORY SYSTEM DEMO")
    print(""*20)

    client = Groq(api_key=api_key)
    memory = MemoryManager(employee_id="EMP001")

    # Preload long-term memory
    memory.long_term = {
        "employee_id": "EMP001",
        "preferred_language": "English"
    }

    demo_exchanges = [
        "My name is Ajay.",
        "What is my name?",
        "I prefer to speak in Hindi.",
        "What language do I prefer?",
        "My employee ID is EMP001.",
        "Can you summarize what you know about me?",
    ]

    print(f"\n{'='*60}")
    print(" CONVERSATION WITH MEMORY")
    print(f"{'='*60}")

    for user_msg in demo_exchanges:
        print(f"\n👤 User: {user_msg}")
        reply = run_agent_turn(user_msg, memory, client)
        print(f" Agent: {reply}")
        print("-"*60)

    # Show memory state
    memory.display_all_memory()

    # Show JSON file
    print("\n LONG-TERM MEMORY FILE (long_term_memory.json):")
    print(json.dumps(load_long_term_memory(), indent=2))

# ============================================================
# 7. INTERACTIVE CHAT
# ============================================================

def interactive_memory_chat(api_key: str, employee_id: str = "EMP001"):
    print("\n" + "="*60)
    print(" AGENT MEMORY SYSTEM — Interactive Mode")
    print("="*60)
    print("Commands: 'memory' = show memory | 'clear' = clear session | 'quit' = exit")
    print("="*60 + "\n")

    client = Groq(api_key=api_key)
    memory = MemoryManager(employee_id=employee_id)

    existing = get_long_term_memory(employee_id)
    if existing:
        memory.long_term = existing
        name = existing.get("name", "")
        if name:
            print(f" Welcome back, {name}! (Loaded from long-term memory)\n")

    while True:
        user_input = input("You: ").strip()
        if not user_input:
            continue
        if user_input.lower() == "quit":
            print("Agent: Goodbye! Your memory has been saved. 👋")
            break
        if user_input.lower() == "memory":
            memory.display_all_memory()
            continue
        if user_input.lower() == "clear":
            memory.short_term.clear()
            print(" Short-term memory cleared. Long-term memory preserved.\n")
            continue

        reply = run_agent_turn(user_input, memory, client)
        print(f"🤖 Agent: {reply}\n")

# ============================================================
#  RUN HERE — Change API key and execute this cell
# ============================================================

API_KEY = "API KEY"   # 🔑 paste your Groq key

run_memory_demo(API_KEY)

# Uncomment for interactive mode:
# interactive_memory_chat(API_KEY, employee_id="EMP001")
````

````python

import json
import time
from groq import Groq

# ============================================================
# MODEL
# ============================================================

MODEL = "llama-3.3-70b-versatile"

# ============================================================
# 1. KNOWLEDGE BASE
# ============================================================

KNOWLEDGE_BASE = {
    "python": "Python is a high-level, interpreted programming language created by Guido van Rossum, first released in 1991.",
    "machine learning": "Machine Learning is a subset of AI that enables systems to learn from data without being explicitly programmed.",
    "deep learning": "Deep Learning is a subset of machine learning using neural networks with many layers to learn representations of data.",
    "llm": "Large Language Models (LLMs) are AI models trained on massive text datasets to understand and generate human language.",
    "groq": "Groq is an AI infrastructure company providing fast LLM inference using their custom LPU (Language Processing Unit) hardware.",
    "india": "India is a country in South Asia with a population of over 1.4 billion, capital New Delhi, and independence on August 15, 1947.",
    "sun": "The Sun is the star at the center of our Solar System, approximately 4.6 billion years old, with surface temperature of ~5,500°C.",
    "water": "Water (H2O) boils at 100°C (212°F) at sea level and freezes at 0°C (32°F).",
    "einstein": "Albert Einstein (1879-1955) was a German-born physicist who developed the theory of relativity and won the Nobel Prize in Physics in 1921.",
}

# ============================================================
# 2. PROMPT BUILDERS
# ============================================================

def build_initial_prompt(user_query: str) -> str:
    return f"""Answer the following question clearly and completely:

Question: {user_query}

Provide a detailed, accurate answer."""

def build_validation_prompt(user_query: str, initial_response: str) -> str:
    return f"""You are a critical reviewer. Evaluate the response below for quality issues.

Original Question: {user_query}

Initial Response:
{initial_response}

Check for ALL of the following issues and list each one found:

1. MISSING INFORMATION - Is any important detail left out?
2. INCORRECT FACTS - Are there any factual errors or inaccuracies?
3. INCOMPLETE ANSWER - Does it fully answer the question?
4. CLARITY ISSUES - Is any part confusing or poorly explained?
5. STRUCTURE ISSUES - Is the answer well-organized?

Respond ONLY in this exact JSON format:
{{
  "issues_found": true or false,
  "missing_information": "describe what is missing or say None",
  "incorrect_facts": "describe any factual errors or say None",
  "incomplete_answer": "describe what is incomplete or say None",
  "clarity_issues": "describe clarity problems or say None",
  "structure_issues": "describe structure problems or say None",
  "overall_score": <score 1-10>,
  "needs_correction": true or false,
  "summary": "one sentence summary of main issues"
}}"""

def build_correction_prompt(user_query: str, initial_response: str, validation_result: dict) -> str:
    return f"""You are an expert writer. Improve the response based on the validation feedback.

Original Question: {user_query}

Initial Response:
{initial_response}

Validation Feedback:
- Missing Information : {validation_result.get('missing_information', 'None')}
- Incorrect Facts     : {validation_result.get('incorrect_facts', 'None')}
- Incomplete Answer   : {validation_result.get('incomplete_answer', 'None')}
- Clarity Issues      : {validation_result.get('clarity_issues', 'None')}
- Structure Issues    : {validation_result.get('structure_issues', 'None')}
- Overall Score       : {validation_result.get('overall_score', 'N/A')}/10
- Summary             : {validation_result.get('summary', '')}

Write an IMPROVED response that:
 Fixes all identified issues
 Adds any missing information
 Corrects any factual errors
 Is clearly structured and complete
 Fully answers the original question

Improved Response:"""

# ============================================================
# 3. REFLECTION ENGINE (Rate-Limit Safe)
# ============================================================

class ReflectionEngine:
    def __init__(self, api_key: str):
        self.client = Groq(api_key=api_key)
        self.model  = MODEL

    def _call_llm(self, prompt: str, temperature: float = 0.7) -> str:
        """Call LLM with retry on rate limit."""
        for attempt in range(3):
            try:
                response = self.client.chat.completions.create(
                    model=self.model,
                    messages=[{"role": "user", "content": prompt}],
                    max_tokens=500,
                    temperature=temperature,
                )
                return response.choices[0].message.content or ""
            except Exception as e:
                if "rate_limit" in str(e).lower() or "429" in str(e):
                    wait = (attempt + 1) * 30
                    print(f"    Rate limit hit. Waiting {wait}s... (attempt {attempt+1}/3)")
                    time.sleep(wait)
                else:
                    raise e
        return "Rate limit exceeded. Please wait and try again."

    def generate_initial_response(self, user_query: str) -> str:
        """Step 1: Generate initial response."""
        print("\n STEP 1: Generating Initial Response...")
        print("-" * 50)
        response = self._call_llm(
            build_initial_prompt(user_query),
            temperature=0.7
        )
        print(response)
        time.sleep(5)
        return response

    def validate_response(self, user_query: str, initial_response: str) -> dict:
        """Step 2: Validate the initial response."""
        print("\n STEP 2: Validating Response...")
        print("-" * 50)
        raw = self._call_llm(
            build_validation_prompt(user_query, initial_response),
            temperature=0.1
        )
        time.sleep(5)

        # Parse JSON safely
        try:
            clean = raw.strip()
            if "```" in clean:
                clean = clean.split("```")[1]
                if clean.startswith("json"):
                    clean = clean[4:]
            result = json.loads(clean.strip())
        except json.JSONDecodeError:
            result = {
                "issues_found"      : True,
                "missing_information": "Could not parse validation",
                "incorrect_facts"   : "None",
                "incomplete_answer" : "None",
                "clarity_issues"    : "None",
                "structure_issues"  : "None",
                "overall_score"     : 5,
                "needs_correction"  : True,
                "summary"           : "Validation parsing failed — applying correction anyway."
            }

        # Display results
        print(f"   Quality Score    : {result.get('overall_score', 'N/A')}/10")
        print(f"   Issues Found    : {result.get('issues_found', False)}")
        print(f"   Missing Info     : {result.get('missing_information', 'None')}")
        print(f"   Incorrect Facts  : {result.get('incorrect_facts', 'None')}")
        print(f"   Incomplete       : {result.get('incomplete_answer', 'None')}")
        print(f"   Clarity Issues   : {result.get('clarity_issues', 'None')}")
        print(f"   Structure Issues : {result.get('structure_issues', 'None')}")
        print(f"   Needs Correction : {result.get('needs_correction', False)}")
        print(f"   Summary          : {result.get('summary', '')}")
        return result

    def self_correct(self, user_query: str, initial_response: str, validation: dict) -> str:
        """Step 3: Generate corrected/improved response."""
        print("\n  STEP 3: Self-Correcting Response...")
        print("-" * 50)
        corrected = self._call_llm(
            build_correction_prompt(user_query, initial_response, validation),
            temperature=0.5
        )
        print(corrected)
        return corrected

    def run(self, user_query: str) -> dict:
        """
        Full pipeline:
        Initial Response → Validation → Self-Correction → Improved Response
        """
        print("\n" + "="*60)
        print(" REFLECTION & SELF-CORRECTION ENGINE")
        print("="*60)
        print(f" Query: {user_query}")
        print("="*60)

        # Step 1
        initial = self.generate_initial_response(user_query)

        # Step 2
        validation = self.validate_response(user_query, initial)

        # Step 3
        score            = validation.get("overall_score", 0)
        needs_correction = validation.get("needs_correction", True)

        if needs_correction or score < 9:
            final     = self.self_correct(user_query, initial, validation)
            corrected = True
        else:
            print("\n Response already high quality. No correction needed.")
            final     = initial
            corrected = False

        # Final summary
        print("\n" + "="*60)
        print(" FINAL IMPROVED RESPONSE")
        print("="*60)
        print(final)
        print("="*60)
        print(f"\n Quality Score : {score}/10")
        print(f" Was Corrected : {corrected}")

        return {
            "query"           : user_query,
            "initial_response": initial,
            "validation"      : validation,
            "final_response"  : final,
            "was_corrected"   : corrected,
            "quality_score"   : score
        }

# ============================================================
# 4. SINGLE QUERY MODE
# ============================================================

def run_single(api_key: str, query: str):
    engine = ReflectionEngine(api_key=api_key)
    return engine.run(query)

# ============================================================
# 5. INTERACTIVE MODE
# ============================================================

def interactive_mode(api_key: str):
    engine = ReflectionEngine(api_key=api_key)
    print("\n" + "="*60)
    print(" REFLECTION ENGINE — Interactive Mode")
    print("="*60)
    print("Type any question. Agent will reflect and self-correct.")
    print("Type 'quit' to exit.\n")

    while True:
        query = input(" Your Question: ").strip()
        if not query:
            continue
        if query.lower() in ["quit", "exit"]:
            print("Goodbye! ")
            break
        engine.run(query)
        print("\n" + "-"*60 + "\n")
        time.sleep(10)   # pause between questions

# ============================================================
#  RUN — Paste API key and execute
# ============================================================

API_KEY = "API KEY"   # 🔑 Your Groq API key

#  Option A: Single query (recommended — saves tokens)
run_single(API_KEY, "What is machine learning?")

# Option B: Interactive mode (uncomment)
# interactive_mode(API_KEY)
````

# Task 8: Build a Multi-Agent System

Building a Multi-Agent System:

A Multi-Agent System (MAS) consists of multiple AI agents working together, where each agent has a specialized role and collaborates to solve complex tasks.

Why Multi-Agent Systems?

A single agent may struggle with:

Large tasks,
Complex reasoning,
Multiple domains,
Long workflows.


Instead, divide responsibilities among specialized agents.

````python


import json
import time
from groq import Groq

# ============================================================
# MODEL
# ============================================================

MODEL = "llama-3.3-70b-versatile"

# ============================================================
# 1. BASE AGENT
# ============================================================

class BaseAgent:
    def __init__(self, name: str, role: str, api_key: str):
        self.name   = name
        self.role   = role
        self.client = Groq(api_key=api_key)

    def _call_llm(self, system_prompt: str, user_prompt: str) -> str:
        """Call LLM with retry on rate limit."""
        for attempt in range(3):
            try:
                response = self.client.chat.completions.create(
                    model=MODEL,
                    messages=[
                        {"role": "system", "content": system_prompt},
                        {"role": "user",   "content": user_prompt}
                    ],
                    max_tokens=800,
                    temperature=0.7,
                )
                return response.choices[0].message.content or ""
            except Exception as e:
                if "rate_limit" in str(e).lower() or "429" in str(e):
                    wait = (attempt + 1) * 30
                    print(f"   ⏳ Rate limit. Waiting {wait}s... (attempt {attempt+1}/3)")
                    time.sleep(wait)
                else:
                    raise e
        return "Rate limit exceeded. Please try again later."

    def run(self, input_data: dict) -> dict:
        raise NotImplementedError("Each agent must implement run()")

    def print_header(self):
        print(f"\n{'='*60}")
        print(f"🤖 {self.name}")
        print(f"   Role: {self.role}")
        print(f"{'='*60}")

# ============================================================
# 2. RESEARCH AGENT
# ============================================================

class ResearchAgent(BaseAgent):
    def __init__(self, api_key: str):
        super().__init__(
            name = "Research Agent",
            role = "Collect and summarize information about the topic",
            api_key = api_key
        )

    def run(self, input_data: dict) -> dict:
        self.print_header()
        user_request = input_data.get("user_request", "")

        system_prompt = """You are an expert Research Agent. Your job is to:
- Collect key information about the requested topic
- Identify market trends and opportunities
- Summarize existing solutions and their limitations
- Highlight key challenges and requirements
Be concise, factual, and structured."""

        user_prompt = f"""Research the following topic thoroughly:

Topic: {user_request}

Provide:
1. Overview of the topic
2. Current market landscape
3. Key challenges & pain points
4. Existing solutions & their gaps
5. Key opportunities
6. Important technical considerations"""

        print("\n📡 Researching topic...")
        result = self._call_llm(system_prompt, user_prompt)
        print(result)
        time.sleep(5)

        return {
            **input_data,
            "research_output": result,
            "agent_trail": input_data.get("agent_trail", []) + ["Research Agent"]
        }

# ============================================================
# 3. BUSINESS ANALYST AGENT
# ============================================================

class BusinessAnalystAgent(BaseAgent):
    def __init__(self, api_key: str):
        super().__init__(
            name = "Business Analyst Agent",
            role = "Generate business requirements from research",
            api_key = api_key
        )

    def run(self, input_data: dict) -> dict:
        self.print_header()
        user_request    = input_data.get("user_request", "")
        research_output = input_data.get("research_output", "")

        system_prompt = """You are an expert Business Analyst. Your job is to:
- Convert research into clear business requirements
- Define functional and non-functional requirements
- Identify stakeholders and their needs
- Define success metrics and KPIs
Be specific, measurable, and business-focused."""

        user_prompt = f"""Based on the research below, generate business requirements.

Original Request: {user_request}

Research Findings:
{research_output}

Generate:
1. Business Objectives
2. Functional Requirements (FR1, FR2, ...)
3. Non-Functional Requirements (NFR1, NFR2, ...)
4. Stakeholders & their needs
5. Success Metrics / KPIs
6. Constraints & Assumptions"""

        print("\n📋 Generating business requirements...")
        result = self._call_llm(system_prompt, user_prompt)
        print(result)
        time.sleep(5)

        return {
            **input_data,
            "ba_output": result,
            "agent_trail": input_data.get("agent_trail", []) + ["BA Agent"]
        }

# ============================================================
# 4. ARCHITECT AGENT
# ============================================================

class ArchitectAgent(BaseAgent):
    def __init__(self, api_key: str):
        super().__init__(
            name = "Architect Agent",
            role = "Generate system design from requirements",
            api_key = api_key
        )

    def run(self, input_data: dict) -> dict:
        self.print_header()
        user_request = input_data.get("user_request", "")
        ba_output    = input_data.get("ba_output", "")

        system_prompt = """You are an expert Software Architect. Your job is to:
- Design scalable and maintainable system architecture
- Select appropriate technologies and frameworks
- Define system components and their interactions
- Address non-functional requirements (performance, security, scalability)
Be technical, precise, and comprehensive."""

        user_prompt = f"""Based on the business requirements below, design the system architecture.

Original Request: {user_request}

Business Requirements:
{ba_output}

Generate:
1. High-Level Architecture (components & interactions)
2. Technology Stack (frontend, backend, database, infra)
3. System Components & their responsibilities
4. Data Flow & API Design
5. Security & Scalability considerations
6. Architecture Diagram (text-based)"""

        print("\n🏗️  Designing system architecture...")
        result = self._call_llm(system_prompt, user_prompt)
        print(result)
        time.sleep(5)

        return {
            **input_data,
            "architecture_output": result,
            "agent_trail": input_data.get("agent_trail", []) + ["Architect Agent"]
        }

# ============================================================
# 5. PROJECT MANAGER AGENT
# ============================================================

class ProjectManagerAgent(BaseAgent):
    def __init__(self, api_key: str):
        super().__init__(
            name = "Project Manager Agent",
            role = "Generate project roadmap and execution plan",
            api_key = api_key
        )

    def run(self, input_data: dict) -> dict:
        self.print_header()
        user_request         = input_data.get("user_request", "")
        ba_output            = input_data.get("ba_output", "")
        architecture_output  = input_data.get("architecture_output", "")

        system_prompt = """You are an expert Project Manager. Your job is to:
- Create realistic project roadmaps and timelines
- Break down work into phases, milestones, and tasks
- Identify risks and mitigation strategies
- Define team structure and resource requirements
Be practical, detailed, and delivery-focused."""

        user_prompt = f"""Based on the requirements and architecture below, generate a project roadmap.

Original Request: {user_request}

Business Requirements Summary:
{ba_output[:500]}...

Architecture Summary:
{architecture_output[:500]}...

Generate:
1. Project Phases & Timeline (Phase 1, 2, 3...)
2. Milestones & Deliverables
3. Team Structure & Roles
4. Sprint Plan (first 3 sprints)
5. Risk Register (Risk + Impact + Mitigation)
6. Budget Estimate (rough)"""

        print("\n📅 Generating project roadmap...")
        result = self._call_llm(system_prompt, user_prompt)
        print(result)
        time.sleep(5)

        return {
            **input_data,
            "pm_output": result,
            "agent_trail": input_data.get("agent_trail", []) + ["PM Agent"]
        }

# ============================================================
# 6. REPORT GENERATOR
# ============================================================

def generate_final_report(pipeline_output: dict) -> str:
    user_request = pipeline_output.get("user_request", "")
    trail        = pipeline_output.get("agent_trail", [])

    report = f"""
{'='*60}
📄 FINAL MULTI-AGENT REPORT
{'='*60}
📌 Project: {user_request}
🤖 Agent Pipeline: {' → '.join(trail)}
{'='*60}

━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
📡 SECTION 1: RESEARCH FINDINGS
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
{pipeline_output.get('research_output', 'N/A')}

━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
📋 SECTION 2: BUSINESS REQUIREMENTS
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
{pipeline_output.get('ba_output', 'N/A')}

━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
🏗️  SECTION 3: SYSTEM ARCHITECTURE
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
{pipeline_output.get('architecture_output', 'N/A')}

━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
📅 SECTION 4: PROJECT ROADMAP
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
{pipeline_output.get('pm_output', 'N/A')}

{'='*60}
✅ REPORT COMPLETE
{'='*60}
"""
    return report

# ============================================================
# 7. MULTI-AGENT ORCHESTRATOR
# ============================================================

class MultiAgentOrchestrator:
    def __init__(self, api_key: str):
        self.api_key  = api_key
        self.agents   = [
            ResearchAgent(api_key),
            BusinessAnalystAgent(api_key),
            ArchitectAgent(api_key),
            ProjectManagerAgent(api_key),
        ]

    def run(self, user_request: str) -> str:
        print("\n" + "🤖 "*20)
        print("     MULTI-AGENT SYSTEM STARTING")
        print("🤖 "*20)
        print(f"\n📌 User Request: {user_request}")
        print(f"\n🔄 Pipeline:")
        print("   User Request → Research → BA → Architect → PM → Final Report")

        # Initial input
        pipeline_data = {
            "user_request": user_request,
            "agent_trail" : []
        }

        # Run each agent sequentially
        for agent in self.agents:
            pipeline_data = agent.run(pipeline_data)

        # Generate final report
        print("\n\n" + "="*60)
        print("📄 GENERATING FINAL REPORT...")
        print("="*60)
        report = generate_final_report(pipeline_data)
        print(report)

        # Save report to file
        filename = "multi_agent_report.txt"
        with open(filename, "w") as f:
            f.write(report)
        print(f"\n💾 Report saved to: {filename}")

        return report

# ============================================================
# ✅ RUN
# ============================================================

API_KEY = "API KEY"   # 🔑 Your Groq API key

orchestrator = MultiAgentOrchestrator(api_key=API_KEY)

# Change this to any project idea!
orchestrator.run("Build an AI-powered HR Management System")
````
